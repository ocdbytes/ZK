## Bilinear Pairings

```sh
f(aG,bG) = f(abG,G) = f(G,abG)
f: G × G → Gᴛ

# also possible : 
e(a, b) → c, a ∈ G, b ∈ G′, c ∈ G′′
```

````python
################################
#     Bilinear Pairings        #
################################

from py_ecc.bn128 import G1, G2, pairing, add, multiply, eq, neg

print(G1)
print(G2)

# Here G1 and G2 both have similar behaviours

assert add(G1, G1) == multiply(G1,2)
assert add(G2, G2) == multiply(G2,2)
````

````python
# Generating the pairing for 5 and 6
A = multiply(G2, 5)
B = multiply(G1, 6)
C = multiply(G2, 5 * 6)

# print(pairing(A, B)) # -> Gives us an output of 12 dimensions -> G12

assert pairing(A, B) == pairing(C, G1)

# MATH : e: G1 × G2 → G12
````
